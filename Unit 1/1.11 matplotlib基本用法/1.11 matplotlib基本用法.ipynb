## 相关知识

matplotlib，是matrix + plot + library的缩写，虽然命名很是直观，但调用它的时候居然还不能简单的直接调用，而是要用它的子模块pyplot。

**简单的线条**

使用import导入模块matplotlib.pyplot，并简写成plt。使用import导入模块numpy，并简写成np

```python
import matplotlib.pyplot as plt
import numpy as np
```

使用np.linspace定义x：范围是(-1,1);个数是50. 仿真一维数据组(x ,y)表示曲线1.

```python
x = np.linspace(-1, 1, 50)
y = 2*x + 1
```

使用plt.figure定义一个图像窗口. 使用plt.plot画(x ,y)曲线. 使用plt.show显示图像.

```python
plt.figure()
plt.plot(x, y)
plt.show()
```

同时，我们也可以根据自己的需要对曲线的样式进行修改。比如说我们将曲线的颜色属性(color)设置为红色;曲线的宽度(linewidth)设置为1.0；曲线的类型(linestyle)设置为虚线.

```python
plt.figure()
plt.plot(x, y, color='red', linewidth=5, linestyle='--')
plt.show()
```

### 设置坐标轴

**调整名字和间隔**

使用plt.xlim设置x坐标轴范围：(-1, 2)；

使用plt.ylim设置y坐标轴范围：(-2, 3)； 

使用plt.xlabel设置x坐标轴名称：'I am x'； 

使用plt.ylabel设置y坐标轴名称：'I am y'；

```python
plt.figure()
plt.plot(x, y, color='red', linewidth=5, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.show()
```

如果标签中含有中文，就会报错。错误根源：Anaconda3的字体库中没有中文字simhei.ttf

所以我们需要在代码中加入这两行代码

plt.rcParams['font.sans-serif']=['SimHei']

plt.rcParams['axes.unicode_minus'] = False

```python
plt.figure()
plt.plot(x, y, color='red', linewidth=5, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.show()
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

使用np.linspace定义范围以及个数：范围是(-1,2);个数是5. 使用print打印出新定义的范围. 使用plt.xticks设置x轴刻度：范围是(-1,2);个数是5.

```python
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.figure()
plt.xticks(new_ticks)
plt.plot(x, y, color='red', linewidth=5, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.show()
```

黑色边界框线在Matplotlib中被称为spines，中文翻译为脊柱

我们要挪动的便是这样四个spine，所有的操作均在plt.gca( )中完成，gca就是get current axes的意思

使用plt.gca获取当前坐标轴信息. 

使用.spines设置边框：右侧边框；

使用.set_color设置边框颜色：默认白色； 

使用.spines设置边框：上边框；

使用.set_color设置边框颜色：默认白色；

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

使用.xaxis.set_ticks_position设置x坐标刻度数字或名称的位置：bottom.（所有位置：top，bottom，both，default，none）

使用.spines设置边框：x轴；

使用.set_position设置边框位置：y=0的位置；（位置所有属性：outward，axes，data）

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
plt.show()
```

使用.yaxis.set_ticks_position设置y坐标刻度数字或名称的位置：left.（所有位置：left，right，both，default，none）

使用.spines设置边框：y轴；

使用.set_position设置边框位置：x=0的位置；（位置所有属性：outward，axes，data） 使用plt.show显示图像.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.show()
plt.show()
```

legend将要显示的信息来自于代码中的 label. 所以我们只需要简单写下一下代码, plt 就能自动的为我们添加图例.

```python
l1, = plt.plot(x, y1, label='linear line')
l2, = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')
plt.legend(loc='lower right')
```

其中'loc'参数有多种，'best'表示自动分配最佳位置，其余的如下：
 'best' : 0,   
 
 'upper right'  : 1,
 
 'upper left'   : 2,
 
 'lower left'   : 3,
 
 'lower right'  : 4,
 
 'right'      : 5,
 
 'center left'  : 6,
 
 'center right' : 7,
 
 'lower center' : 8,
 
 'upper center' : 9,
 
 'center'     : 10,

### 参考答案

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(-5, 4, 30)
m1 = 3 * n + 2
m2 = n ** 2

plt.plot(n, m1, 'r-.', n, m2, 'b')

plt.xlim((-2, 4))
plt.ylim((-5, 15))

x_ticks = np.linspace(-4, 4, 9)
plt.xticks(x_ticks)

plt.yticks([-10,10], ['hate','love'])

plt.xlabel('X')
plt.ylabel('Y')

ax = plt.gca()

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

plt.show()
```
